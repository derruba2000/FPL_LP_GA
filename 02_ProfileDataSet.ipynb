```

from pandas_profiling import ProfileReport
import pandas as pd
import numpy as np


def ProfileGW(DataFolder):
  df = pd.read_parquet("{0}/GWSummary.parquet".format(DataFolder))
  df=df[df["total_points"]>0] # Filter players who have not scored any point
  df_profile = ProfileReport(df)
  df_profile.to_file(f"{DataFolder}/DataProfile.html")
  return df_profile
```
